Map of kenya and Uganda

```python
import cartopy
import numpy as np
import pandas as pd
import geopandas
import contextily as cx
import cartopy.crs as ccrs
import matplotlib.pyplot as plt
```

```python
from cartopy.io import shapereader

# get country borders
resolution = '10m'
category = 'cultural'
name = 'admin_0_countries'

shpfilename = shapereader.natural_earth(resolution, category, name)

# read the shapefile using geopandas
df = geopandas.read_file(shpfilename)
```

```python
countries = df.query("ADMIN in ['Uganda', 'Kenya']")
```

```python
borders = df.query("ADMIN in ['Burundi', 'Rwanda','Democratic Republic of the Congo', 'United Republic of Tanzania', 'Ethiopia', 'South Sudan', 'Somalia']")
```

```python
lat, long = 0.46588, 34.09221

kis_lat, kis_long = -0.091702, 34.76795
```

```python
import cartopy
import seaborn as sns
```

```python
ymin, ymax, xmin, xmax = 6, 6, 29, 42
```

```python
fig, ax = plt.subplots(1,1, figsize=[10,10])

countries.plot(ax=ax, facecolor='none', edgecolor='k', linewidth=1, alpha=0.25)
cx.add_basemap(ax, crs=df.crs.to_string(), source=cx.providers.Stamen.TerrainBackground)
borders.plot(ax=ax, facecolor='w', edgecolor='w', linewidth=1)
ax.plot(long, lat, markersize=15, marker='o', markerfacecolor='w', markeredgecolor='k', alpha=0.6)
ax.text(x=long-0.5, y=lat+0.3, s="Busia", fontdict={'fontsize':16})

ax.plot(kis_long, kis_lat, markersize=15, marker='o', markerfacecolor='w', markeredgecolor='k', alpha=0.6)
ax.text(x=kis_long-0.58, y=kis_lat-0.45, s="Kisumu", fontdict={'fontsize':16})
ax.text(x=32, y=1.75, s="Uganda", fontdict={'fontsize':20, 'weight':'bold'})
ax.text(x=37, y=0, s="Kenya", fontdict={'fontsize':20, 'weight':'bold'})
fig.savefig("Uganda_Kenya.svg", dpi=300)
```
